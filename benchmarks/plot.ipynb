```julia
using CSV
using DataFrames
using Plots
using StatsPlots
     
```

```julia
"Read and concatenate multiple data frames from files"
function read_and_concatenate(fnames...)
    df = CSV.read(fnames[1], DataFrame)
    
    for fname ∈ fnames[2:end]
        df2 = CSV.read(fname, DataFrame)
        df = [df; df2]
    end
    df
end
```

```julia
julia_n2plain = read_and_concatenate("./results/time_julia_plain.csv")
```

```julia
julia_n2tiled = read_and_concatenate("./results/time_julia_tiled.csv")
```

```julia
python_n2plain = read_and_concatenate("./results/time_python_plain.csv")
```

```julia
python_n2tiled = read_and_concatenate("./results/time_python_tiled.csv")
```

```julia
@df julia_n2plain plot(:mean_particles, :time_per_event, label="", xlabel="<n> initial particles", ylabel="Jet Reconstruction Time μs/event")

@df julia_n2plain scatter!(:mean_particles, :time_per_event, label = "Julia N2Plain")
@df julia_n2tiled plot!(:mean_particles, :time_per_event, label="")
@df julia_n2tiled scatter!(:mean_particles, :time_per_event, label = "Julia N2Tiled")

@df python_n2plain plot!(:mean_particles, :time_per_event, label="")
@df python_n2plain scatter!(:mean_particles, :time_per_event, label = "python N2Plain")
@df python_n2tiled plot!(:mean_particles, :time_per_event, label="")
@df python_n2tiled scatter!(:mean_particles, :time_per_event, label = "python N2Tiled")
```
